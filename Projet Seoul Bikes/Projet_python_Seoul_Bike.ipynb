```python
# Encoding categorical variables (we use get_dummies rather than LabelEncoder to avoid introducing any notion of order)
data = pd.get_dummies(data, columns=['Seasons', 'Holiday', 'Day', 'Month'])
data
```

```python
data.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, LogisticRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

X = data.drop('Rented Bike Count', axis=1)
y = data['Rented Bike Count']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

models_performance = {}

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)
models_performance['Linear Regression'] = {'MSE': mse_lr, 'R2': r2_lr}

# Logistic Regression
# logistic = LogisticRegression()
# logistic.fit(X_train, y_train)
# y_pred_logistic = logistic.predict(X_test)
# mse_logistic = mean_squared_error(y_test, y_pred_logistic)
# r2_logistic = r2_score(y_test, y_pred_logistic)
# models_performance['Logistic Regression'] = {'MSE': mse_logistic, 'R2': r2_logistic}

# Ridge Regression
ridge = Ridge(random_state=42)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
models_performance['Ridge Regression'] = {'MSE': mse_ridge, 'R2': r2_ridge}

# Lasso Regression
lasso = Lasso(random_state=42)
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)
models_performance['Lasso Regression'] = {'MSE': mse_lasso, 'R2': r2_lasso}

# Decision Tree Classifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
y_pred_dt_classifier = dt_classifier.predict(X_test)
mse_dt_classifier = mean_squared_error(y_test, y_pred_dt_classifier)
r2_dt_classifier = r2_score(y_test, y_pred_dt_classifier)
models_performance['Decision Tree Classifier'] = {'MSE': mse_dt_classifier, 'R2': r2_dt_classifier}

# Random Forest Classifier
# rf_classifier = RandomForestClassifier(random_state=42)
# rf_classifier.fit(X_train, y_train)
# y_pred_rf_classifier = rf_classifier.predict(X_test)
# mse_rf_classifier = mean_squared_error(y_test, y_pred_rf_classifier)
# r2_rf_classifier = r2_score(y_test, y_pred_rf_classifier)
# models_performance['Random Forest Classifier'] = {'MSE': mse_rf_classifier, 'R2': r2_rf_classifier}

# K-Nearest Neighbors Classifier
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)
y_pred_knn_classifier = knn_classifier.predict(X_test)
mse_knn_classifier = mean_squared_error(y_test, y_pred_knn_classifier)
r2_knn_classifier = r2_score(y_test, y_pred_knn_classifier)
models_performance['K-Nearest Neighbors Classifier'] = {'MSE': mse_knn_classifier, 'R2': r2_knn_classifier}

# Random Forest Regressor
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
models_performance['Random Forest Regressor'] = {'MSE': mse_rf, 'R2': r2_rf}

# Gradient Boosting Regressor
gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
y_pred_gbr = gbr.predict(X_test)
mse_gbr = mean_squared_error(y_test, y_pred_gbr)
r2_gbr = r2_score(y_test, y_pred_gbr)
models_performance['Gradient Boosting Regressor'] = {'MSE': mse_gbr, 'R2': r2_gbr}

# Find the best performing model based on R2 score
best_model = max(models_performance, key=lambda x: models_performance[x]['R2'])

models_performance
```

```python
# Charger le modèle
rf_model = joblib.load('rf_manual_model.joblib')

# Données d'entrée modifiées
input_data = {
    'Hour': [13],  # Heure de la journée (13h)
    'Temperature(°C)': [22.3],  # Température en Celsius
    'Humidity(%)': [56],  # Pourcentage d'humidité
    'Wind speed (m/s)': [2.3],  # Vitesse du vent (m/s)
    'Visibility (10m)': [2000],  # Visibilité (10m)
    'Dew point temperature(°C)': [14.2],  # Température du point de rosée
    'Solar Radiation (MJ/m2)': [0.5],  # Radiation solaire (MJ/m2)
    'Rainfall(mm)': [0],  # Précipitations (mm)
    'Snowfall (cm)': [0],  # Chutes de neige (cm)
    'Seasons': ['Autumn'],  # Saisons 
    'Holiday': ['Holyday'],  # Vacances 
    'Functioning Day': ['Yes'],  # Jour de fonctionnement 
    'Date': ['01/02/2023']  # Date 
}

# Transformer 'Date' en 'Day' et 'Month'
input_df = pd.DataFrame(input_data)
input_df['Date'] = pd.to_datetime(input_df['Date'], format='%d/%m/%Y')
input_df['Day'] = input_df['Date'].dt.day_name()
input_df['Month'] = input_df['Date'].dt.month_name()

# Clé pour appliquer la prédiction seulement si le shop est ouvert
key = input_df['Functioning Day'][0] == 'Yes'

input_df.drop(['Date', 'Functioning Day'], axis=1, inplace=True)
    
# Encodage des variables catégorielles
input_df_encoded = pd.get_dummies(input_df, drop_first=False)

# Ajouter manuellement les colonnes manquantes et les initialiser à 0
expected_features = [
    'Hour', 'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)', 
    'Visibility (10m)', 'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)', 
    'Rainfall(mm)', 'Snowfall (cm)', 'Seasons_Autumn', 'Seasons_Spring', 
    'Seasons_Summer', 'Seasons_Winter', 'Holiday_Holiday', 'Holiday_No Holiday', 'Day_Friday', 'Day_Monday', 
    'Day_Saturday', 'Day_Sunday', 'Day_Thursday', 'Day_Tuesday', 'Day_Wednesday',
    'Month_April', 'Month_August', 'Month_December', 'Month_February', 
    'Month_January', 'Month_July', 'Month_June', 'Month_March', 'Month_May', 
    'Month_November', 'Month_October', 'Month_September'
]

for feature in expected_features:
    if feature not in input_df_encoded.columns:
        input_df_encoded[feature] = 0

# Assurer l'ordre correct des colonnes
input_df_encoded = input_df_encoded[expected_features]

# Prédiction
if key:
    prediction = rf_model.predict(input_df_encoded)
    print(prediction)
else:
    print("The shop is closed")
```

```python
from flask import Flask, request, send_from_directory


app = Flask(__name__)


@app.route('/')
def index():
    return send_from_directory('static', 'index.html')


@app.post('/predict')
def predict():
    form_data = request.form.to_dict()
    
    # Transformer 'Date' en 'Day' et 'Month'
    form_data_lists = {k: [v] for k, v in form_data.items()}
    input_df = pd.DataFrame(form_data_lists)

    # Conversion des colonnes numériques
    numeric_columns = ['Hour', 'Temperature(°C)', 'Humidity(%)', 'Wind speed (m/s)', 'Visibility (10m)', 'Dew point temperature(°C)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)']
    for col in numeric_columns:
        input_df[col] = pd.to_numeric(input_df[col])

    input_df['Date'] = pd.to_datetime(input_df['Date'], format='%d/%m/%Y')
    input_df['Day'] = input_df['Date'].dt.day_name()
    input_df['Month'] = input_df['Date'].dt.month_name()

    key = input_df['Functioning Day'][0] == 'Yes'

    input_df.drop(['Date', 'Functioning Day'], axis=1, inplace=True)

    input_df_encoded = pd.get_dummies(input_df, drop_first=False)

    # Ajout des caractéristiques catégorielles manquantes
    for feature in expected_features:
        if feature not in input_df_encoded.columns:
            input_df_encoded[feature] = 0

    input_df_encoded = input_df_encoded[expected_features]

    if key:
        prediction = rf_model.predict(input_df_encoded)
        return {"prediction": prediction.tolist()}
    else:
        return {"prediction": "The shop is closed"}


if __name__ == '__main__':
    app.run()
```
